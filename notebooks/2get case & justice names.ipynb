```python
def get_case_name(txts):
    #gets case name (title) from full text. probably could do this a simpler way...
    name = re.compile(r'United States Supreme Court (.+?)(?=No)')
    casename = name.findall(txts)
    
    try:
        casename = casename[0].strip()

    except IndexError:
        if re.search(r'[\(\)\d]+',txts):
            print('found')
            nlist= re.compile('[\(\)\d]+').split(txts)
            name = nlist[0]
            casename = name.strip().strip(',') 
        #return name
    
    if re.search(r'[\(\)\d]+',txts):
        nlist= re.compile('[\(\)\d]+').split(txts)
        casename = nlist[0]
    try:
        if re.search('United States Supreme Court', casename):
            nlist = re.compile('United States Supreme Court').split(casename)
            casename = nlist[1]
            casename = casename.strip().strip(',')
    except TypeError:
        print()
    try:
        if re.search('FindLaw: US Supreme Court Opinions:', casename):
            nlist = re.compile('FindLaw: US Supreme Court Opinions:').split(casename)

            casename = nlist[1].strip()
    except TypeError:
        print()
    
    print(casename)
    try:
        return casename.strip().strip(',')
    except AttributeError:
        return casename
```

```python
court_text['case_name'] = court_text.fulltext.apply(get_case_name)
#court_text[court_text.case_name.map(len)<200]
#
```

```python
#get justice names out of document

def get_justice(split_on):
    
    if re.search(r"(?i)(?<=justice\s)[a-z\']*\b", split_on):
        name = re.search(r"(?i)(?<=justice\s)[a-z\']*\b", split_on).group(0)
        print(name)
        return name
```

```python
justice_op = pd.DataFrame(columns = ['justice', 'year', 'case', 'text'])

count = 0
for index, row in court_text.iterrows():
    txt = row.fulltext
    year = row.year
    case = row.case_name
    
    #remove anything after footnotes (dont want it)
    if re.search("(?i)FOOTNOTES", txt):
        textlist = txt.split("(?i)FOOTNOTES")
        txt = textlist[0]
        
    #split on this phrase as it means a new justice/opinion
    if re.search("on writ of certiorari to", txt):
        oplist = txt.split("on writ of certiorari to")
        oplist.pop(0)
        for i in oplist:
            name = re.search(r"(?i)(?<=justice\s)[a-z\']*\b", i)
            if name:
                justice = name.group(0)
                justice_op.loc[count] = [justice, year, case, i]
                count+=1
                
    #docs before 1996 or so dont have that phrase, need to do some regex to get the different opinions            
    else:
        
        if re.search("(?i)JUSTICE (.+?)(?<=delivered the opinion of the Court)", txt):
            majname = re.search("(?i)JUSTICE (.+?)(?<=delivered the opinion of the Court)", txt).group(0)
            tlist = txt.split(majname)
            tlist.pop(0)
            justice = get_justice(majname)
            print(justice)
            op = tlist[0]
            if re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)", op):
                pattern = re.compile("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)")
                op = pattern.split(op)
                op = op[0]
            if re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=dissenting)", op):
                pattern = re.compile("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=dissenting)")
                op = pattern.split(op)
                op = op[0] 
            
            justice_op.loc[count] = [justice, year, case, op]
            count+=1
        
        if re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)", txt):
            concname = re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)", txt).group(0)
            tlist = txt.split(concname)
            tlist.pop(0)
            op = tlist[0]
            cjustice = get_justice(concname)
            if re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)", op):
                pattern = re.compile("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=concurring)")
                op = pattern.split(op)
                op = op[0]
            justice_op.loc[count] = [cjustice, year, case, op]
            count+=1
        if re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=dissenting)", txt):
            disname = re.search("(?=.{3,200}$)(?i)JUSTICE (.+?)(?<=dissenting)", txt).group(0)
            djustice = get_justice(disname)
            tlist = txt.split(disname)
            tlist.pop(0)
            op = tlist[0]
            justice_op.loc[count] = [djustice, year, case, op]
            count+=1

```
